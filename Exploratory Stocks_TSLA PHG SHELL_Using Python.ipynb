```python
shel_df['Ticker'] = "SHEL"
phg_df['Ticker'] = "PHG"
tsla_df['Ticker'] = "TSLA"
df = pd.concat([shel_df, phg_df, tsla_df])
df = df[['Ticker', 'Returns']]
print(df.head())
```

```python
sns.boxplot(x= df['Ticker'], y= df['Returns'])
plt.title('Box Plot for SHEL, PHG, and TSLA Returns')
plt.show()
```

```python
df_corr = pd.DataFrame({'SHEL': shel_df['Returns'], 'PHG': phg_df['Returns'], 'TSLA': tsla_df['Returns']})
df_corr
```

```python
corr = df_corr.corr()
corr
```

```python
sns.heatmap(corr, cmap='RdBu', vmin=-1, vmax=1, annot = True, square=True, 
            annot_kws={'fontsize':12, 'fontweight': 'bold'})
plt.rcParams['figure.figsize'] = (10,8)
plt.show()
```

```python
import datetime
```

```python
cutoff = datetime.datetime(2023,1,1)
shel_df['Date'] = pd.to_datetime(shel_df['Date'], format='%Y/%m/%d')
shel_df = shel_df[shel_df['Date'] > cutoff]
shel_df['SMA_10'] = shel_df['Close'].rolling(window=10).mean()
print(shel_df.head())
plt.plot(shel_df['Date'], shel_df['SMA_10'])
plt.plot(shel_df['Date'], shel_df['Adj Close'])
plt.title("Moving average and Adj Close price for SHEL")
plt.ylabel('Adj Close Price')
plt.xlabel('Date')
plt.show()
```

```python
cutoff = datetime.datetime(2023,1,1)
phg_df['Date'] = pd.to_datetime(phg_df['Date'], format='%Y/%m/%d')
phg_df = phg_df[phg_df['Date'] > cutoff]
phg_df['SMA_10'] = phg_df['Close'].rolling(window=10).mean()
print(phg_df.head())
plt.plot(phg_df['Date'], phg_df['SMA_10'])
plt.plot(phg_df['Date'], phg_df['Adj Close'])
plt.title("Moving average and Adj Close price for PHG")
plt.ylabel('Adj Close Price')
plt.xlabel('Date')
plt.show()
```

```python
cutoff = datetime.datetime(2023,1,1)
tsla_df['Date'] = pd.to_datetime(tsla_df['Date'], format='%Y/%m/%d')
tsla_df = tsla_df[tsla_df['Date'] > cutoff]
tsla_df['SMA_10'] = tsla_df['Close'].rolling(window=10).mean()
print(tsla_df.head())
plt.plot(tsla_df['Date'], tsla_df['SMA_10'])
plt.plot(tsla_df['Date'], tsla_df['Adj Close'])
plt.title("Moving average and Adj Close price for TSLA")
plt.ylabel('Adj Close Price')
plt.xlabel('Date')
plt.show()
```

```python
shel_df['SMA_10_STD'] = shel_df['Adj Close'].rolling(window=20).std()
shel_df['Upper Band'] = shel_df['SMA_10'] + (shel_df['SMA_10_STD'] * 2)
shel_df['Lower Band'] = shel_df['SMA_10'] - (shel_df['SMA_10_STD'] * 2)
shel_df.index = shel_df['Date']
shel_df[['Adj Close', 'SMA_10', 'Upper Band', 'Lower Band']].plot(figsize=(12,6))
plt.title('10 Day Bollinger Band for Shell')
plt.ylabel('Adjusted Close Price')
plt.show()
```

```python
phg_df['SMA_10_STD'] = phg_df['Adj Close'].rolling(window=20).std()
phg_df['Upper Band'] = phg_df['SMA_10'] + (phg_df['SMA_10_STD'] * 2)
phg_df['Lower Band'] = phg_df['SMA_10'] - (phg_df['SMA_10_STD'] * 2)
phg_df.index = phg_df['Date']
phg_df[['Adj Close', 'SMA_10', 'Upper Band', 'Lower Band']].plot(figsize=(12,6))
plt.title('10 Day Bollinger Band for Philips')
plt.ylabel('Adjusted Close Price')
plt.show()
```
