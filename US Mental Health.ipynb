

#### All values from 3 columns are skewed, so using Median to fill missing data

### Fill NaN values for DatValue, LowConfidenceLimit and HighConfidenceLimit by Median

```python
# Fill NaN values for DatValue by MEDIAN

median_datavalue = Mental_Health['DataValue'].astype('float').median(axis=0)
print('Median of DataValue is: ', median_datavalue)
Mental_Health['DataValue'].replace(np.nan, median_datavalue, inplace=True)

# Fill NaN values for LowConfidenceLimit by Median

median_LowCo = Mental_Health['LowConfidenceLimit'].astype('float').median(axis=0)
print('Median of Low Confidence Limit is: ', median_LowCo)
Mental_Health['LowConfidenceLimit'].replace(np.nan, median_LowCo, inplace=True)

# Fill NaN values for HighConfidenceLimit by Median

median_HighCo = Mental_Health['HighConfidenceLimit'].astype('float').median(axis=0)
print('Median of High Confidence Limit is: ', median_HighCo)
Mental_Health['HighConfidenceLimit'].replace(np.nan, median_HighCo, inplace=True)
```

```python
### Check NaN values for DatValue, LowConfidenceLimit and HighConfidenceLimit

print('Null DataValue:', Mental_Health['DataValue'].isnull().sum())
print('Null LowConfidenceLimit:', Mental_Health['LowConfidenceLimit'].isnull().sum())
print('Null HighConfidenceLimit:', Mental_Health['HighConfidenceLimit'].isnull().sum())
```

```python
# Convert 'DataValue' to float
Mental_Health['DataValue'] = Mental_Health['DataValue'].astype('float')

# check the data type of 'DataValue'
print(Mental_Health['DataValue'].dtype)
```

## Outlier Detection

```python
# Find out the Quanrtile Range for columns 'DataValue'
Q1_DV = Mental_Health['DataValue'].quantile(.25)
Q3_DV = Mental_Health['DataValue'].quantile(.75)
IQR_DV = Q1_DV - Q3_DV
print('The Inter Quantile Range for DataValue:', IQR_DV)

# Find out the Quanrtile Range for columns 'HighConfidenceLimit'
Q1_HC = Mental_Health['HighConfidenceLimit'].quantile(.25)
Q3_HC = Mental_Health['HighConfidenceLimit'].quantile(.75)
IQR_HC = Q1_HC - Q3_HC
print('The Inter Quantile Range for HighConfidenceLimit:', IQR_HC)
```

```python
# Find out uper and lowe bounds
upper_DV = Q3_DV+(IQR_DV*1.5)
lower_DV = Q1_DV+(IQR_DV*1.5)

upper_HC = Q3_HC+(IQR_HC*1.5)
lower_HC = Q1_HC+(IQR_HC*1.5)

print('Upper Bound DataValue:', upper_DV)
print('Lower Bound DataValue:', lower_DV)
print('Upper HighConfidenceLimit:', upper_HC)
print('Lower HighConfidenceLimit:', lower_HC)
```

```python
# Removing Outlier

Mental_Health['DataValue'].clip(upper_DV, lower_DV)
Mental_Health['HighConfidenceLimit'].clip(upper_HC, lower_HC)

Mental_Health[['DataValue','LowConfidenceLimit','HighConfidenceLimit']].describe()
```

# 5. Data Preparation

```python

# Filter the dataset for the year 2021
Mental_Health_2021 = Mental_Health[Mental_Health['Years'] == 2021]

# Create data from top 10 countries with highest mental illness percentange

states = Mental_Health.groupby('States')['DataValue'].mean().reset_index()
states = states.sort_values(by='DataValue', ascending=False).head(10)

# Create new features of prevalence for each group
prevalence_over_years = Mental_Health.groupby(['Years','StratificationCategory'])['DataValue'].mean().reset_index()

# Collect neccessary columns
West_Virginia = Mental_Health[Mental_Health['States'] == 'West Virginia']
West_Virginia_by_years = West_Virginia.groupby('Years')['DataValue'].mean().reset_index()
US = Mental_Health.groupby('Years')['DataValue'].mean().reset_index()

# Fill data in West_Virginia and the US for only 2018 and 2021
West_Virginia_18_21 = West_Virginia[West_Virginia['Years'].isin ([2018, 2021])]
US_18_21 = Mental_Health[Mental_Health['Years'].isin([2018, 2021])]

West_Virginia_18_21 = West_Virginia_18_21.groupby(['Years', 'StratificationCategory'])['DataValue'].mean().unstack().reset_index()
US_18_21 = US_18_21.groupby(['Years', 'StratificationCategory'])['DataValue'].mean().unstack().reset_index()

# Filter the dataset for the year 2018
Mental_Health_2018 = Mental_Health[Mental_Health['Years'] == 2018]




```
